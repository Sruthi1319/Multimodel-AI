```python
import gradio as gr
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
import torch


#Load BLIP model and processor
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large")
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

#Define captioning Function
def generate_caption(image):
  inputs = processor(images=image, return_tensors="pt").to(device)
  output = model.generate(**inputs,max_new_tokens=20)
  caption = processor.decode(output[0], skip_special_tokens=True)
  return caption

#Gradio UI
gr.Interface(
    fn=generate_caption,
  inputs=gr.Image(type="pil",label="Upload an Image"),
  outputs=gr.Textbox(label="Generated Caption "),
  title="Image Captioning with BLIP",
  description="Open-source Multimodal:Describe images with natural lanaguage"
).launch()

```

```python
#2. Visual questioning and answer
import gradio as gr
from PIL import Image
from transformers import BlipProcessor, BlipForQuestionAnswering
import torch

#Using BLIP for simplicity,fast + open source
processor = BlipProcessor.from_pretrained("Salesforce/blip-vqa-base")
model = BlipForQuestionAnswering.from_pretrained("Salesforce/blip-vqa-base")
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

#Function to answer visual questions
def vqa_answer(image,question):
  inputs = processor(images=image, text=question, return_tensors="pt").to(device)
  output = model.generate(**inputs)
  answer = processor.decode(output[0], skip_special_tokens=True)
  return answer

#gardio UI
gr.Interface(
    fn=vqa_answer,
    inputs=[
        gr.Image(type="pil", label="Upload an Image"),
        gr.Textbox(label="Ask a question about the image")
    ],
    outputs=gr.Textbox(label="Answer"),
    title="Visual Question Answering with BLIP",
    description="Ask a question about an image using BLIP."
).launch()

```

```python
import gradio as gr
from PIL import Image
from transformers import BlipProcessor, BlipForQuestionAnswering
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
import torch

# Load BLIP and MBART
blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-vqa-base")
blip_model = BlipForQuestionAnswering.from_pretrained("Salesforce/blip-vqa-base")

mbart_model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
mbart_tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")

# Device setup
device = "cuda" if torch.cuda.is_available() else "cpu"
blip_model.to(device)
mbart_model.to(device)

# Language dropdown options
language_options = {
    "English": "en_XX",
    "Hindi": "hi_IN",
    "Telugu": "te_IN",
    "French": "fr_XX",
    "German": "de_DE"
}

# Translate English answer to selected language
def translate(answer_en, lang_code):
    if lang_code == "en_XX":
        return answer_en  # no translation needed

    mbart_tokenizer.src_lang = "en_XX"
    encoded = mbart_tokenizer(answer_en, return_tensors="pt").to(device)
    generated = mbart_model.generate(
        **encoded,
        forced_bos_token_id=mbart_tokenizer.lang_code_to_id[lang_code]
    )
    return mbart_tokenizer.decode(generated[0], skip_special_tokens=True)

# Main function
def vqa_with_translation(image, question, target_language):
    # Get English answer using BLIP
    inputs = blip_processor(images=image, text=question, return_tensors="pt").to(device)
    output = blip_model.generate(**inputs)
    answer_en = blip_processor.decode(output[0], skip_special_tokens=True)

    # Translate if needed
    lang_code = language_options[target_language]
    final_answer = translate(answer_en, lang_code)

    return final_answer

# Gradio UI
gr.Interface(
    fn=vqa_with_translation,
    inputs=[
        gr.Image(type="pil", label="üì∑ Upload an Image"),
        gr.Textbox(label="‚ùì Ask a question about the image"),
        gr.Dropdown(label="üåê Choose Output Language", choices=list(language_options.keys()), value="English")
    ],
    outputs=gr.Textbox(label="üìù Answer"),
    title="üåç Multilingual Visual Question Answering with Translator",
    description="Upload an image, ask a question, and choose the language you want the answer in (English, Hindi, Telugu, French, German)."
).launch()
```

```python
import os
import cv2
import gradio as gr
from PIL import Image
from transformers import BlipProcessor, BlipForQuestionAnswering, BlipForConditionalGeneration
import torch
#Load BLIP model
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
#Extract frames from video (1frame per second)
def extract_frames(video_path,fps_step=1):
  vidcap = cv2.VideoCapture(video_path)
  fps = int(vidcap.get(cv2.CAP_PROP_FPS))
  success,image = vidcap.read()
  count = 0
  frames = []
  while success:
    if count % (fps *fps_step) == 0:
      frame_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
      pil_image = Image.fromarray(frame_rgb)
      frames.append(pil_image)
    success,frame = vidcap.read()
    count += 1
    vidcap.release()
  return frames

#Caption generator
def generate_video_captions(video):
  #Get the actual file path from gradio upload
  if hasattr(video,'name'):
    video_path = video.name
  elif isinstance(video,str):
    video_path = video
  else:
    return "Could not read video file."
  if not os.path.exists(video_path):
    return "Video file not found."

  #Extract and caption
  frames = extract_frames(video_path,fps_step=1)
  if not frames:
    return "No frames extracted from the video."
    captions = []
  for i,img in enumerate(frames):
    inputs = processor(images=img,return_tensors="pt").to(device)
    out = model.generate(**inputs,max_new_tokens=50)
    caption = processor.decode(out[0],skip_special_tokens=True)
    caption.append(f"Frame {i+1}: {caption}")
  return "\n".join(caption)
  #Gradio UI (No type="filepath")
  gr.Interface(
      fn=generate_video_captions,
      inputs=gr.Video(label="Upload a Video"),
      outputs=gr.Textbox(label="Generated Captions"),
      title="Video Captioning with BLIP",
      description="Extracts 1 frame per second from a video and generates captions using BLIP.",
  ).launch()
```
